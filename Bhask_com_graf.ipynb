```python
from sympy import *
from sympy.plotting import *

y = Function('y')
x = Symbol('x')

y = x**2 - 6*x + 5 
 
solve(y, x) #Raiz da equação 2° grau
```

```python
y1 = diff(y, x)
x_v = solve(y1, x)  # O numero 3 é o x do vertice   
y_v = y.subs(x, 3)  # Torcando de x para 3
print('O ponto do vertice é: {}, {}'.format(x_v, y_v))
```

```python
plot(y, xlim=[0,8], ylim=[-5, 5])
```
